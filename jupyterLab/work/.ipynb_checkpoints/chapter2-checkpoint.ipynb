# 簡単な線形回帰をしてみよう

```python
import numpy as np
import matplotlib.pyplot as plt
import torch
from torchviz import make_dot
import japanize_matplotlib
```

## サンプルデータの入力

```python
smapleDara1 = np.array([
    [166,58.7],
    [176.0,75.7],
    [171.0,62.1],
    [173.0,70.4],
    [169.0,60.1]
])

x=smapleDara1[:,0]
y=smapleDara1[:,1]
```

```python
plt.grid(True)
plt.scatter(x,y,c='k',s=50)
plt.xlabel('$x$: 身長(cm)')
plt.ylabel('$y$: 体重(kg)')
plt.title('身長と体重の関係')
plt.show()
```

## データの変換

```python
X = x - x.mean()
Y = y - y.mean()
plt.grid(True)
plt.scatter(X,Y,c='k',s=50)
plt.xlabel('$X$: 身長(cm)')
plt.ylabel('$Y$: 体重(kg)')
plt.title('加工後の身長と体重の関係')
plt.show()
```

## テンソルに変換

```python
X = torch.tensor(X).float()
Y = torch.tensor(Y).float()
```

## パラメータの定義

```python
W = torch.tensor(1.0,requires_grad=True).float()
B = torch.tensor(1.0,requires_grad=True).float()
```

## 予測関数の定義

```python
def pred(X):
    return W*X+B
```

```python
Yp=pred(X)
```

```python
params = {'W':W,'B':B} 
g=make_dot(Yp,params=params)
display(g)
```

## 損失関数の定義

```python
def mse(Yp,Y):
    loss=((Yp-Y)**2).mean()
    return loss
```

```python
loss=mse(Yp,Y)
params = {'W':W,'B':B} 
g=make_dot(loss,params=params)
display(g)
```

```python
loss.backward()
```

## パラメータの更新

```python
lr=0.001
with torch.no_grad():
    W-=lr*W.grad
    B-=lr*B.grad
    W.grad.zero_()
    B.grad.zero_()
```
